```python
# ============================================
# RAW LAYER — Load Superstore CSV and 
# save as Parquet files
# Author: Sadhana S Kumar
# ============================================

import pandas as pd
import os
from datetime import datetime

# ── Define paths ──────────────────────────────
BASE_PATH    = r"C:\Users\Lenovo\OneDrive\Desktop\medallion-project"
RAW_PATH     = os.path.join(BASE_PATH, "raw")
SILVER_PATH  = os.path.join(BASE_PATH, "silver")
GOLD_PATH    = os.path.join(BASE_PATH, "gold")

# ── Load Superstore CSV ───────────────────────
print("📥 Loading Superstore CSV...")

# Change this path to where your Superstore file is!
CSV_PATH = r"C:\Users\Lenovo\OneDrive\Desktop\dataset\archive\Sample - Superstore.csv"

df = pd.read_csv(CSV_PATH, encoding='latin1')

print(f"✅ Loaded {len(df)} rows and {len(df.columns)} columns")
print(f"\n📋 Columns: {list(df.columns)}")
print(f"\n🔍 Sample data:")
df.head()
```

```python
# ============================================
# Save Raw data as Parquet files
# This is our RAW LAYER in Medallion Architecture
# ============================================

from datetime import datetime

# Today's date for partitioning
load_date = datetime.today().strftime('%Y-%m-%d')

# ── Save full load as parquet ─────────────────
full_load_path = os.path.join(RAW_PATH, f"load_date={load_date}")
os.makedirs(full_load_path, exist_ok=True)

df.to_parquet(
    os.path.join(full_load_path, "superstore.parquet"),
    index=False
)

print(f"✅ Raw layer created!")
print(f"📁 Location: {full_load_path}")
print(f"📦 Format: Parquet")
print(f"🔢 Rows saved: {len(df)}")
print(f"\n💡 This is your RAW LAYER in Medallion Architecture!")
print(f"   Just like Oracle DB → ADLS in your real project!")
```

```python
# ============================================
# SILVER LAYER — Read Parquet from Raw,
# Clean data and save to Silver
# Just like Databricks reads from ADLS!
# ============================================

print("📥 Reading from Raw layer (Parquet)...")

# Read parquet file — just like Databricks reads from ADLS
silver_df = pd.read_parquet(
    os.path.join(RAW_PATH, f"load_date={load_date}", 
                 "superstore.parquet")
)

print(f"✅ Read {len(silver_df)} rows from Raw layer")

# ── Clean & Transform ─────────────────────────
print("\n🧹 Cleaning data...")

# 1. Fix date columns
silver_df['Order Date'] = pd.to_datetime(silver_df['Order Date'])
silver_df['Ship Date']  = pd.to_datetime(silver_df['Ship Date'])

# 2. Remove duplicates
before = len(silver_df)
silver_df = silver_df.drop_duplicates(subset=['Order ID', 'Product ID'])
after = len(silver_df)
print(f"   Removed {before - after} duplicate rows")

# 3. Remove nulls
silver_df = silver_df.dropna(subset=['Sales', 'Profit'])
print(f"   Removed null rows")

# 4. Add audit columns — just like in real Databricks!
silver_df['load_date']      = load_date
silver_df['load_timestamp'] = datetime.now()
silver_df['source_system']  = 'Superstore_CSV'
silver_df['load_type']      = 'full_load'

print(f"\n✅ Silver layer has {len(silver_df)} clean rows")

# ── Save to Silver layer ──────────────────────
silver_path = os.path.join(SILVER_PATH, "superstore")
os.makedirs(silver_path, exist_ok=True)

silver_df.to_parquet(
    os.path.join(silver_path, "superstore_silver.parquet"),
    index=False
)

print(f"\n✅ Silver layer created!")
print(f"📁 Location: {silver_path}")
print(f"🔢 Rows saved: {len(silver_df)}")
print(f"\n💡 This is your SILVER LAYER!")
print(f"   Just like Delta tables in Databricks!")
```

```python
# ============================================
# GOLD LAYER — Read from Silver,
# Apply business transformations
# Ready for Power BI consumption!
# ============================================

print("📥 Reading from Silver layer...")

gold_df = pd.read_parquet(
    os.path.join(SILVER_PATH, "superstore", 
                 "superstore_silver.parquet")
)

print(f"✅ Read {len(gold_df)} rows from Silver layer")

# ── Apply Business Transformations ────────────
print("\n⚡ Applying transformations for Power BI team...")

# 1. Extract date parts
gold_df['year']    = gold_df['Order Date'].dt.year
gold_df['month']   = gold_df['Order Date'].dt.month
gold_df['quarter'] = gold_df['Order Date'].dt.quarter
gold_df['month_name'] = gold_df['Order Date'].dt.strftime('%B')

# 2. Calculate KPIs
gold_df['profit_margin_pct'] = round(
    gold_df['Profit'] / gold_df['Sales'] * 100, 2
)
gold_df['revenue_after_discount'] = round(
    gold_df['Sales'] * (1 - gold_df['Discount']), 2
)
gold_df['shipping_days'] = (
    gold_df['Ship Date'] - gold_df['Order Date']
).dt.days

# 3. Add customer segment flag
gold_df['is_enterprise'] = gold_df['Segment'].apply(
    lambda x: True if x == 'Corporate' else False
)

# 4. Drop audit columns — Gold is clean for PBI!
gold_df = gold_df.drop(columns=[
    'load_date', 'load_timestamp', 
    'source_system', 'load_type'
])

print(f"✅ Transformations applied!")
print(f"\n📊 New columns added for Power BI:")
print(f"   ✅ year, month, quarter, month_name")
print(f"   ✅ profit_margin_pct")
print(f"   ✅ revenue_after_discount")
print(f"   ✅ shipping_days")
print(f"   ✅ is_enterprise")

# ── Save to Gold layer ────────────────────────
gold_path = os.path.join(GOLD_PATH, "superstore")
os.makedirs(gold_path, exist_ok=True)

gold_df.to_parquet(
    os.path.join(gold_path, "superstore_gold.parquet"),
    index=False
)

print(f"\n✅ Gold layer created!")
print(f"📁 Location: {gold_path}")
print(f"🔢 Rows saved: {len(gold_df)}")
print(f"\n💡 This is your GOLD LAYER!")
print(f"   Ready for Power BI to connect and build dashboards!")
```
